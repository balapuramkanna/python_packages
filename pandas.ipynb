# **clean the null values**

  - they are 3 types
  - 1. dropping missing values(dropna())
     - it is the way to deal with missing data to drop it
     - dropna() is used for this method
     - we can using this method entire null data will loss

```python
import pandas as pd
df = pd.read_csv(r"/content/sample_data/Diabetes Missing Data.csv")
df.head(10)
```

```python
# finding null values
df.isnull().sum()
```

```python
#shape
print(df.shape)
#using dropna method
df1 = df.dropna()
print(df1.isnull().sum())
df1.shape
```

```python
#using inplace method
import pandas as pd
df = pd.read_csv(r"/content/sample_data/Diabetes Missing Data.csv")
df.head()
print(df.shape)
df.dropna(inplace = True , axis = 0)
print(df.isnull().sum())
print(df.shape)
```

2.replacing missing values:
 - we can repace the values with mean or median
 - we can call it as impitation or replacement techinque

```python
import pandas as pd
df = pd.read_csv(r"/content/sample_data/Diabetes Missing Data.csv")
print(df.isnull().sum())
df.shape
```

```python
# replacing null values using mean

mean_value = df.mean()
print(mean_value)
df1 = df.fillna(mean_value)
df1.head()
```

```python
# using median
median_value = df.median()
print(median_value)
df2 = df.fillna(median_value)
print(median_value)
print(df2.shape)
df2.head()
```

3.duplicate data:
  - it will drop the repeating values
  - it will checks row by row if there is any repeating values it can drop

```python
import pandas as pd
df = pd.read_csv(r"/content/sample_data/Diabetes Missing Data.csv")
print(df.isnull().sum())
print(df.shape)
df1 = df.drop_duplicates()
df1.shape
```

# Aggregating data with groupby() in pandas:
  
  - it is used for summarinzing the data aggregating more then 2 columns.
  - aggregative known as sum and average.
  - when we apply grouby method we need to clean the data



```python
import pandas as pd
df = pd.read_csv(r"/content/sample_data/Diabetes Missing Data.csv")

df.head()
# clean the data
median_value = df.median()
print(median_value)
df1 = df.fillna(median_value).head(20)
df1.head(20)
df1.isnull().sum()
df1.groupby("Glucose").median().head()
```
